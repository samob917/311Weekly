```python
def build_prompt():
    complaint =""
    for i in df['complaint_type'].head(50):
        complaint=complaint+i
    print(complaint)
    return complaint
    

def generate_phrase(prompt:str):
    response = openai.Completion.create(engine="text-davinci-003",
                                      prompt=prompt,
                                      max_tokens=1000,
                                      temperature=0.8)
  #sleep(5)
  #print(response)
    response["choices"][0]["text"]
    return response


place = "NEW YORK"

csv_output = all_cities[place][["complaint_type", "city","status"]][:20].to_csv()
csv_output
prompt = build_prompt()
summary = generate_phrase(f"""See complaints from residents of {place}. Can you summarize these complaints and format as a newsletter, say what the top 4 complaints with how much complaints there are: {csv_output}""")
summary['choices'][0]['text']
```

```python
after_4 = False
summary_string = ""
for line in text.split('\n'):
    if after_4:
        summary_string += line
    if "1." in line:
        complaint_1 = line.split(".")[1].split("(")[0].strip()
        num_complaint_1 = line.split(".")[1].split("(")[1].split(" ")[0]
        summary_string += line.split(":")[1]
    elif "2." in line:
        complaint_2 = line.split(".")[1].split("(")[0].strip()
        num_complaint_2 = line.split(".")[1].split("(")[1].split(" ")[0]
        summary_string += line.split(":")[1]
    elif "3." in line:
        complaint_3 = line.split(".")[1].split("(")[0].strip()
        num_complaint_3 = line.split(".")[1].split("(")[1].split(" ")[0]
        summary_string += line.split(":")[1]
    elif "4." in line:
        complaint_4 = line.split(".")[1].split("(")[0].strip()
        num_complaint_4 = line.split(".")[1].split("(")[1].split(" ")[0]
        summary_string += line.split(":")[1]
        after_4 = True
print(summary_string)
    
```

```python
from jinja2 import Environment, FileSystemLoader

# Create a Jinja environment and load the template
env = Environment(loader=FileSystemLoader('.'))
template = env.get_template('email_template.html')

# Define context data
context = {'city': place, 'week_of': week_of, 'complaint_1': complaint_1, "num_complaint_1": num_complaint_1, 'complaint_2': complaint_2, "num_complaint_2": num_complaint_2, 'complaint_3': complaint_3, "num_complaint_3": num_complaint_3, 'complaint_4': complaint_4, "num_complaint_4": num_complaint_4, "general_summary": summary_string, "num_solved": closed, "num_in_progress": in_progress }

# Render the template with the context
output = template.render(context)

output_file_path = "output.html"

with open(output_file_path, 'w') as output_file:
    output_file.write(output)
# Print or use the output as needed
print(output)
```
